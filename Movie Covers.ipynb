```python
import csv
import urllib.parse
import urllib.request
from bs4 import BeautifulSoup
```

```python
with open('/dataset/movie_poster.csv', 'r', newline='', encoding = 'utf8') as in_csv:
    reader = csv.DictReader(in_csv, delimiter=',')
    rows = list(reader)
    row_atual = len(rows) - 1
    print(rows[row_atual]['movieId'])

#with open('./dataset/movie_poster.csv', 'w', newline='') as out_csv:
#    writer = csv.writer(out_csv, delimiter=',')
#    writer.writerow(['movieId', 'image'])

with open('./dataset/links.csv', 'r', newline='', encoding = 'utf8') as in_csv:
    reader = csv.DictReader(in_csv, delimiter=',')
    for row in reader:
        if int(row['movieId']) > int(rows[row_atual]['movieId']):
            movie_id = row['movieId']
            imdb_id = row['imdbId']
            domain = 'https://www.imdb.com/title/tt'
            try:
                url = domain + imdb_id
                with urllib.request.urlopen(url) as response:
                    html = response.read()
                    soup = BeautifulSoup(html, 'html.parser')
                    # Get url of poster image
                    try:
                        image_url = soup.find('div', class_='poster').a.img['src']
                        print(image_url)
                        with open('./dataset/movie_poster.csv', 'a', newline='') as out_csv:
                            writer = csv.writer(out_csv, delimiter=',')
                            writer.writerow([movie_id, image_url])
                    # Ignore cases where search returns no results
                    except AttributeError:
                        pass
            except urllib.error.HTTPError:
                pass
```
